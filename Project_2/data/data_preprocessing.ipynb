```python
# opening neccessary csv files
life_satisfaction = pd.read_csv('life_satisfaction/gdp-vs-happiness.csv')
ghi = pd.read_csv('life_satisfaction/global-hunger-index.csv')
life_expectancy = pd.read_csv('life_satisfaction/life-satisfaction-vs-life-expectancy.csv')
unemployment = pd.read_csv('life_satisfaction/unemployment-rate.csv')
hdi = pd.read_csv('life_satisfaction/human-development-index.csv')
corruption = pd.read_csv('life_satisfaction/TI-corruption-perception-index.csv')
depression = pd.read_csv('life_satisfaction/depression-vs-self-reported-life-satisfaction.csv')
obesity = pd.read_csv('life_satisfaction/share-of-adults-defined-as-obese.csv')
under_five_deaths = pd.read_csv('life_satisfaction/correlation-between-child-mortality-and-mean-years-of-schooling-for-those-aged-15-and-older.csv')
suicide = pd.read_csv('life_satisfaction/suicide-rate.csv')
fruit_consumption = pd.read_csv('life_satisfaction/fruit-consumption-per-capita.csv')
vegetable_consumption = pd.read_csv('life_satisfaction/vegetable-consumption-per-capita.csv')
birth_rate = pd.read_csv('life_satisfaction/crude-birth-rate.csv')
```

```python
# choosing only data for 2016
life_satisfaction = life_satisfaction[life_satisfaction["Year"]==2016].drop(columns=["Code","Total population (Gapminder, HYDE & UN)",'Continent'])

# dropping NaN values
life_satisfaction = life_satisfaction.dropna(axis=0)

# choosing only developing countries
life_satisfaction_bool = life_satisfaction["Entity"].isin(developing)
filtered_life_satisfaction = life_satisfaction[life_satisfaction_bool]

# renaming columns 
filtered_life_satisfaction = filtered_life_satisfaction.rename(columns={'Entity': 'Country', 
                                                                        'GDP per capita, PPP (constant 2011 international $)': 'GDP',
                                                                         'Life satisfaction in Cantril Ladder (World Happiness Report 2019)':'life_satisfaction'})

# reseting index order 
filtered_life_satisfaction = filtered_life_satisfaction.reset_index(drop=True)
```

```python
# life_satisfaction + GDP
filtered_life_satisfaction
```

```python
# choosing those developing countries that are present in filtered_life_satisfaction dataset
present_developing = filtered_life_satisfaction["Country"].values
```

```python
# choosing only data for 2016
ghi = ghi[ghi["Year"]==2016].drop(columns=["Code"])

# dropping NaN values
ghi = ghi.dropna(axis=0)

# choosing only developing countries
ghi_bool = ghi["Entity"].isin(present_developing)
filtered_ghi = ghi[ghi_bool]

# renaming columns 
filtered_ghi = filtered_ghi.rename(columns={'Entity':'Country',
                                            'Global Hunger Index (IFPRI (2016))':'GHI'})

# reseting index order 
filtered_ghi = filtered_ghi.reset_index(drop=True)

# dropping "Year" column
filtered_ghi = filtered_ghi.drop(columns=["Year"])
```

```python
filtered_ghi
```

```python
# dataset: life_satisfaction+GDP+GHI(Global Hunger Index)
dataset = pd.merge(filtered_life_satisfaction,filtered_ghi,on='Country',how='left')
```
